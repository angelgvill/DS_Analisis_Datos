Aún tengo 292 apariciones así que voy a refinar la selección, si contienen una palabra que sea barrio o distrito.

```python
# Función para verificar coincidencias parciales de 'adress' con los nombres de los barrios
def verificar_coincidencia(address, barrios):
    for barrio in barrios:
        if barrio in address or address in barrio:
            return True
    return False

# Aplicar la función a cada valor de 'address'
barrios_lista = bM['NOMBRE'].tolist()  # Convertir la columna barrios a lista
df['coincide'] = df['address'].apply(lambda x: verificar_coincidencia(x, barrios_lista))

# Comprobar si todos los valores tienen coincidencia
todos_coinciden = df['coincide'].all()

# Resultados
if todos_coinciden:
    print("Todos los valores de 'address' tienen coincidencia parcial con algún barrio.")
else:
    print("Algunos valores de 'address' no tienen coincidencia parcial con ningún barrio.")
```

```python
valores_faltantes_barrio = df[~df['coincide']]
print("Filas cuya 'address' que no incluyen en su texto un barrio de Madrid:")
valores_faltantes_barrio
```

```python
# Función para verificar coincidencias parciales de 'address' con los nombres de los distritos
def verificar_coincidencia(address, distritos):
    for distrito in distritos:
        if distrito in address or address in distrito:
            return True
    return False

# Aplicar la función a cada valor de 'address'
distritos_lista = dM['NOMBRE'].tolist()  # Convertir la columna distritos a lista
valores_faltantes_barrio['coincide'] = valores_faltantes_barrio['address'].apply(lambda x: verificar_coincidencia(x, distritos_lista))

# Comprobar si todos los valores tienen coincidencia
todos_coinciden = valores_faltantes_barrio['coincide'].all()

# Resultados
if todos_coinciden:
    print("Todos los valores de 'address' tienen coincidencia parcial con algún distrito.")
else:
    print("Algunos valores de 'address' no tienen coincidencia parcial con ningún distrito.")
```

```python
# Función para encontrar coincidencia de address con barrio e incorporarlo a una nueva columna 
# en caso de no existir, colocaremos 'ND'
def obtener_barrio(address, barrios):
    for barrio in barrios:
        if barrio in address or address in barrio:
            return barrio
    return "ND"

# Crear nueva columna con el resultado
barrios_lista = bM['NOMBRE'].tolist()
df2['Barrio'] = df2['address'].apply(lambda x: obtener_barrio(x, barrios_lista))

# Mover la columna justo despues de Address
col = df2.pop('Barrio')
df2.insert (9, 'Barrio',col)

df2["Barrio"] = df2["Barrio"].astype("string")
df2
```

Hago lo mismo si coincide con el nombre de distrito y lo incorporo en una nueva columna

```python
# Función para encontrar coincidencia de address con distritos e incorporarlo a una nueva columna 
# y en el caso de no coincidir, buscaremos aqué distrito pertenece ese barrio
def obtener_distrito(address, distritos):
    for distrito in distritos:
        if distrito in address or address in distrito:
            return distrito
    return "completar"

# Crear nueva columna con el resultado
distritos_lista = dM['NOMBRE'].tolist()
df2['Distrito'] = df2['address'].apply(lambda x: obtener_distrito(x, distritos_lista))

# Mover la columna justo despues de Address
col = df2.pop('Distrito')
df2.insert (10, 'Distrito',col)

df2["Distrito"] = df2["Distrito"].astype("string")

df2
```

```python
# Función para buscar el distrito correspondiente al barrio en bM
def obtener_distrito(barrio, bM):
    # Buscar el distrito en el dataframe bM
    distrito = bM.loc[bM['NOMBRE'] == barrio, 'NOMDIS']
    if not distrito.empty:
        return distrito.iloc[0]
    return None  # Si no se encuentra, devolver None

# Aplicar la función solo a las filas donde 'distrito' está vacío
df2['Distrito'] = df2.apply(
    lambda row: obtener_distrito(row['Barrio'], bM) if row['Distrito'] == "completar" else row['Distrito'],
    axis=1
)

df2
```
